```python
import pandas as pd
import numpy as np
import os

# Récupère le dossier où se trouve ce notebook
NOTEBOOK_DIR = os.getcwd()

# '..' remonte d'un cran pour accéder au dossier data, puis sous dossiers Insee et elections
RAW_INSEE_DIR = os.path.join(NOTEBOOK_DIR, '..', 'data', 'raw_Insee')
RAW_ELEC_DIR = os.path.join(NOTEBOOK_DIR, '..', 'data', 'raw_elections')
# On remplace 'processed' par 'merged'
MERGED_DATA_DIR = os.path.join(NOTEBOOK_DIR, '..', 'data', 'merged')

# On vérifie que le dossier de sortie existe, sinon on le crée
os.makedirs(MERGED_DATA_DIR, exist_ok=True)

print(f"OK Source INSEE : {RAW_INSEE_DIR}")
print(f"OK Source Élections : {RAW_ELEC_DIR}")
print(f"OK Destination : {MERGED_DATA_DIR}")
```

```python
# Dictionnaire des fichiers raw INSEE
FILES_INSEE = {
    'pop_2020': 'population_2020.CSV',
    'pop_2014': 'population_2014.xls',
    'diplo_2020': 'diplome_2020.CSV',
    'diplo_2014': 'diplome_2014.xls',
    'rev_2019': 'revenu_2019.csv',
    'rev_2013': 'revenu_2013.xls'
}



# POPULATION & CADRES
# rq: en-têtes pour excel Insee (header=5)
df_pop_2020 = pd.read_csv(os.path.join(RAW_INSEE_DIR, FILES_INSEE['pop_2020']), sep=';', dtype={'COM': str})
df_pop_2014 = pd.read_excel(os.path.join(RAW_INSEE_DIR, FILES_INSEE['pop_2014']), header=5, dtype={'COM': str})

# DIPLÔMES
df_diplo_2020 = pd.read_csv(os.path.join(RAW_INSEE_DIR, FILES_INSEE['diplo_2020']), sep=';', header=0, dtype={'COM': str})
df_diplo_2014 = pd.read_excel(os.path.join(RAW_INSEE_DIR, FILES_INSEE['diplo_2014']), header=5, dtype={'CODGEO': str})
# REVENUS
df_rev_2019 = pd.read_csv(os.path.join(RAW_INSEE_DIR, FILES_INSEE['rev_2019']), sep=';', dtype={'CODGEO': str})
df_rev_2013 = pd.read_excel(os.path.join(RAW_INSEE_DIR, FILES_INSEE['rev_2013']), header=5, dtype={'CODGEO': str})

print("Données chargées.")
```

```python
# Transformation des données socio-démographiques

# Population & cadres
# 2020
pop20 = df_pop_2020[['COM', 'P20_POP', 'C20_POP15P_CS3']].copy()
pop20.columns = ['COM', 'pop_2020', 'cadres_nb_2020']

# 2014 (Attention INSEE change parfois COM en CODGEO, ici on assure)
pop14 = df_pop_2014[['COM', 'P14_POP', 'C14_POP15P_CS3']].copy()
pop14.columns = ['COM', 'pop_2014', 'cadres_nb_2014']

# Diplomes

# 2020 : Il faut sommer les colonnes (Sup court + Sup long...) pour se rabaisser à la definition de 2014 qui est vulgairement agrégée
df_diplo_2020['sup_nb_2020'] = (
    df_diplo_2020['P20_NSCOL15P_SUP2'] + 
    df_diplo_2020['P20_NSCOL15P_SUP34'] + 
    df_diplo_2020['P20_NSCOL15P_SUP5']
)
dip20 = df_diplo_2020[['COM', 'sup_nb_2020']].copy()

# 2014
dip14 = df_diplo_2014[['CODGEO', 'P14_NSCOL15P_SUP']].copy()
dip14.columns = ['COM', 'sup_nb_2014']

# Revenus
rev19 = df_rev_2019[['CODGEO', 'MED19']].rename(columns={'CODGEO': 'COM', 'MED19': 'med_2019'})
rev13 = df_rev_2013[['CODGEO', 'MED13']].rename(columns={'CODGEO': 'COM', 'MED13': 'med_2013'})

# Fusion globale des données socio-démographiques

# On part de la population 2020 comme base
df_socio = pop20.merge(pop14, on='COM', how='inner')
df_socio = df_socio.merge(dip20, on='COM', how='left')
df_socio = df_socio.merge(dip14, on='COM', how='left')
df_socio = df_socio.merge(rev19, on='COM', how='left')
df_socio = df_socio.merge(rev13, on='COM', how='left')

# Calcul des variables Delta

# Croissance Pop (Log-diff)
df_socio['croissance_pop'] = np.log(df_socio['pop_2020']) - np.log(df_socio['pop_2014'])

#  Delta Diplômés (Variation en  %)
df_socio['pct_sup_2020'] = df_socio['sup_nb_2020'] / df_socio['pop_2020']
df_socio['pct_sup_2014'] = df_socio['sup_nb_2014'] / df_socio['pop_2014']
df_socio['delta_pct_diplome'] = df_socio['pct_sup_2020'] - df_socio['pct_sup_2014']

# Delta Cadres (Variation  %)
df_socio['pct_cadres_2020'] = df_socio['cadres_nb_2020'] / df_socio['pop_2020']
df_socio['pct_cadres_2014'] = df_socio['cadres_nb_2014'] / df_socio['pop_2014']
df_socio['delta_pct_cadres'] = df_socio['pct_cadres_2020'] - df_socio['pct_cadres_2014']

# Richesse initiale (Log)
df_socio['log_med_13'] = np.log(df_socio['med_2013'])

print(f"Base socio-démographique prête : {df_socio.shape[0]} communes.")
```

```python
# CHARGEMENT ELECTIONS


# VERIFIER bien CODE NUANCE et pas NUANCE LISTE
df_elec_14 = pd.read_csv(os.path.join(RAW_ELEC_DIR, 'elec_2014.csv'), dtype={'Code commune': str})
df_elec_20 = pd.read_csv(os.path.join(RAW_ELEC_DIR, 'elec_2020.csv'), dtype={'Code commune': str})

# Renommer si nécessaire (verifier Code Nuance)
df_elec_14 = df_elec_14.rename(columns={'Code commune': 'COM'})
df_elec_20 = df_elec_20.rename(columns={'Code commune': 'COM'})

# DÉFINITION DU BLOC GAUCHE + ÉCOLOS 

BLOC_GAUCHE = [
    'LVECE', 'LVEC', 'LECO',         # Écolos
    'LUG', 'LDVG', 'LSOC', 'LFI',    # Gauche diverse, PS, LFI
    'LGEN', 'LCOM', 'LRDG', 'LGRA', 'LPG' # PCF, Front de gauche, etc.
]

# Idée: mettr à 0 le score si pas de liste de gauche. C'est fait plus bas lors du merge

def calcul_score_bloc(df_elec, nuances_bloc, annee):
    """Calcule la somme des voix du bloc pour chaque commune."""
    # Filtrer les nuances du bloc (gauche dans notre cas)
    mask = df_elec['Code Nuance'].isin(nuances_bloc)
    df_filtered = df_elec[mask].copy()
    
    # Sommer par commune (cas où il y a >=2 listes de gauche)
    scores = df_filtered.groupby('COM')['% Voix/Exp'].sum().reset_index()
    scores.rename(columns={'% Voix/Exp': f'score_gauche_{annee}'}, inplace=True)
    return scores

# appel a la fction, calcul des scores
score_14 = calcul_score_bloc(df_elec_14, BLOC_GAUCHE, '2014')
score_20 = calcul_score_bloc(df_elec_20, BLOC_GAUCHE, '2020')

print("Scores calculés.")
```

```python
# On part de la base socio-demo (qui contient toutes les communes)
df_final = df_socio.copy()

# On colle les scores 2014
df_final = df_final.merge(score_14, on='COM', how='left')

# On colle les scores 2020
df_final = df_final.merge(score_20, on='COM', how='left')

# CRUCIAL pour la regression: Remplacer les NaN par 0
# Si pas de correspondance, ça veut dire 0 voix pour la gauche
df_final['score_gauche_2014'] = df_final['score_gauche_2014'].fillna(0)
df_final['score_gauche_2020'] = df_final['score_gauche_2020'].fillna(0)

# Calcul du DELTA SCORE (Variable Y)
df_final['delta_score_gauche'] = df_final['score_gauche_2020'] - df_final['score_gauche_2014']

# Sélection des colonnes (pour R par exemple, analyse faite en R temporairement)
cols_export = [
    'COM',
    'score_gauche_2014',  
    'score_gauche_2020', 
    'delta_score_gauche',
    'pop_2014',
    'pop_2020', 
    'croissance_pop', 
    'pct_sup_2014',
    'pct_sup_2020',
    'delta_pct_diplome', 
    'pct_cadres_2014',
    'pct_cadres_2020',  
    'delta_pct_cadres', 
    'med_2013',
    'med_2019',   
    'log_med_13',
    'log_med_19' 

]

# On retire les lignes avec des NaN dans les variables explicatives (Socio-demo manquantes)
# faire analyse des données perdues ? 
df_export = df_final[cols_export].dropna()

print(f"Base finale prête : {len(df_export)} communes.")
print(df_export.head())
```
