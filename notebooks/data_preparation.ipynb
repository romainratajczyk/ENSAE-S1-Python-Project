```python
import pandas as pd
import numpy as np
import os

# Récupère le dossier où se trouve ce notebook
NOTEBOOK_DIR = os.getcwd()

# '..' remonte d'un cran pour accéder au dossier data, puis sous dossiers Insee et elections
RAW_INSEE_DIR = os.path.join(NOTEBOOK_DIR, '..', 'data', 'raw_Insee')
RAW_ELEC_DIR = os.path.join(NOTEBOOK_DIR, '..', 'data', 'raw_elections')
# On remplace 'processed' par 'merged'
MERGED_DATA_DIR = os.path.join(NOTEBOOK_DIR, '..', 'data', 'merged')

# On vérifie que le dossier de sortie existe, sinon on le crée
os.makedirs(MERGED_DATA_DIR, exist_ok=True)

print(f"OK Source INSEE : {RAW_INSEE_DIR}")
print(f"OK Source Élections : {RAW_ELEC_DIR}")
print(f"OK Destination : {MERGED_DATA_DIR}")
```

```python
# Transformation des données socio-démographiques

# Population & cadres
# 2020
# On groupe par COM et on somme pour éviter les doublons (cas où le fichier source est segmenté par âge/sexe)
pop20 = df_pop_2020.groupby('COM')[['P20_POP', 'C20_POP15P_CS3']].sum().reset_index()
pop20.columns = ['COM', 'pop_2020', 'cadres_nb_2020']

pop22 = df_pop_2022.groupby('COM')[['P22_POP', 'C22_POP15P_STAT_GSEC13_23']].sum().reset_index()
pop22.columns = ['COM', 'pop_2022', 'cadres_nb_2022']

pop17 = df_pop_2017.groupby('COM')[['P17_POP', 'C17_POP15P_CS3']].sum().reset_index()
pop17.columns = ['COM', 'pop_2017', 'cadres_nb_2017']

# 2014 (Attention INSEE change parfois COM en CODGEO, ici on assure)
pop14 = df_pop_2014.groupby('COM')[['P14_POP', 'C14_POP15P_CS3']].sum().reset_index()
pop14.columns = ['COM', 'pop_2014', 'cadres_nb_2014']

# Diplomes

# 2020 : Il faut sommer les colonnes (Sup court + Sup long...) pour se rabaisser à la definition de 2014 qui est vulgairement agrégée
df_diplo_2020['sup_nb_2020'] = (
    df_diplo_2020['P20_NSCOL15P_SUP2'] + 
    df_diplo_2020['P20_NSCOL15P_SUP34'] + 
    df_diplo_2020['P20_NSCOL15P_SUP5']
)
# On groupe aussi ici pour avoir une ligne par commune
dip20 = df_diplo_2020.groupby('COM')['sup_nb_2020'].sum().reset_index()

#2022 
df_diplo_2022['sup_nb_2022'] = (
    df_diplo_2022['P22_NSCOL15P_SUP2'] + 
    df_diplo_2022['P22_NSCOL15P_SUP34'] + 
    df_diplo_2022['P22_NSCOL15P_SUP5']
)
dip22 = df_diplo_2022.groupby('CODGEO')['sup_nb_2022'].sum().reset_index()
dip22.columns = ['COM', 'sup_nb_2022']

# 2017 
df_diplo_2017['sup_nb_2017'] = (
    df_diplo_2017['P17_NSCOL15P_SUP2'] + 
    df_diplo_2017['P17_NSCOL15P_SUP34'] + 
    df_diplo_2017['P17_NSCOL15P_SUP5']
)
dip17 = df_diplo_2017.groupby('CODGEO')['sup_nb_2017'].sum().reset_index()
dip17.columns = ['COM', 'sup_nb_2017']

# 2014
dip14 = df_diplo_2014.groupby('CODGEO')['P14_NSCOL15P_SUP'].sum().reset_index()
dip14.columns = ['COM', 'sup_nb_2014']

# Revenus
# On s'assure d'avoir des valeurs numériques et une seule ligne par commune (drop_duplicates) car on ne somme pas des médianes
df_rev_2019['MED19'] = pd.to_numeric(df_rev_2019['MED19'], errors='coerce')
rev19 = df_rev_2019[['CODGEO', 'MED19']].drop_duplicates(subset='CODGEO').rename(columns={'CODGEO': 'COM', 'MED19': 'med_2019'})

df_rev_2016['MED16'] = pd.to_numeric(df_rev_2016['MED16'], errors='coerce')
rev16 = df_rev_2016[['CODGEO', 'MED16']].drop_duplicates(subset='CODGEO').rename(columns={'CODGEO': 'COM', 'MED16': 'med_2016'})

df_rev_2013['MED13'] = pd.to_numeric(df_rev_2013['MED13'], errors='coerce')
rev13 = df_rev_2013[['CODGEO', 'MED13']].drop_duplicates(subset='CODGEO').rename(columns={'CODGEO': 'COM', 'MED13': 'med_2013'})

# Fusion globale des données socio-démographiques

# On part de la population 2020 comme base
df_socio = pop20.merge(pop14, on='COM', how='inner')
df_socio = df_socio.merge(pop22, on='COM', how='left')
df_socio = df_socio.merge(pop17, on='COM', how='left')
df_socio = df_socio.merge(dip22, on='COM', how='left')
df_socio = df_socio.merge(dip17, on='COM', how='left')
df_socio = df_socio.merge(dip20, on='COM', how='left')
df_socio = df_socio.merge(dip14, on='COM', how='left')
df_socio = df_socio.merge(rev19, on='COM', how='left')
df_socio = df_socio.merge(rev16, on='COM', how='left')
df_socio = df_socio.merge(rev13, on='COM', how='left')

# problème possible log(0): on résout, ainsi que secret statistique et NaN pour les revenus

# Transforme les 's' (secret statistique) en NaN (déjà géré plus haut pour être sûr, mais on garde ta logique)
df_socio['med_2013'] = pd.to_numeric(df_socio['med_2013'], errors='coerce')
df_socio['med_2019'] = pd.to_numeric(df_socio['med_2019'], errors='coerce')
df_socio['med_2016'] = pd.to_numeric(df_socio['med_2016'], errors='coerce')

#  Supprime les lignes avec un revenu ou une population de 0
df_socio = df_socio[(df_socio['med_2013'] > 0) & (df_socio['med_2016'] > 0) & (df_socio['med_2019'] > 0) & (df_socio['pop_2014'] > 0) & (df_socio['pop_2020'] > 0) & (df_socio['pop_2017'] > 0) & (df_socio['pop_2022'] > 0)]


# Calcul des variables Delta

# Croissance Pop (Log-diff)
df_socio['croissance_pop_14_20'] = np.log(df_socio['pop_2020']) - np.log(df_socio['pop_2014'])
# Croissance Pop 2017-2022 (Log-diff)
df_socio['croissance_pop_17_22'] = np.log(df_socio['pop_2022']) - np.log(df_socio['pop_2017'])

#  Delta Diplômés (Variation en  %)
df_socio['pct_sup_2022'] = df_socio['sup_nb_2022'] / df_socio['pop_2022']
df_socio['pct_sup_2020'] = df_socio['sup_nb_2020'] / df_socio['pop_2020']
df_socio['pct_sup_2017'] = df_socio['sup_nb_2017'] / df_socio['pop_2017']
df_socio['delta_pct_diplome_17_22'] = df_socio['pct_sup_2022'] - df_socio['pct_sup_2017']
df_socio['pct_sup_2014'] = df_socio['sup_nb_2014'] / df_socio['pop_2014']
df_socio['delta_pct_diplome_14_20'] = df_socio['pct_sup_2020'] - df_socio['pct_sup_2014']

# Delta Cadres (Variation  %) (On oublie, diplomes racontent globalement la même chose... mais on le met quand même)
df_socio['pct_cadres_2022'] = df_socio['cadres_nb_2022'] / df_socio['pop_2022']
df_socio['pct_cadres_2017'] = df_socio['cadres_nb_2017'] / df_socio['pop_2017']
df_socio['delta_pct_cadres_17_22'] = df_socio['pct_cadres_2022'] - df_socio['pct_cadres_2017']
df_socio['pct_cadres_2020'] = df_socio['cadres_nb_2020'] / df_socio['pop_2020']
df_socio['pct_cadres_2014'] = df_socio['cadres_nb_2014'] / df_socio['pop_2014']
df_socio['delta_pct_cadres_14_20'] = df_socio['pct_cadres_2020'] - df_socio['pct_cadres_2014']

# Richesse initiale (Log)
df_socio['log_med_13'] = np.log(df_socio['med_2013'])

 # richesse en 2016
df_socio['log_med_16'] = np.log(df_socio['med_2016'])

# Richesse en 2019
df_socio['log_med_19'] = np.log(df_socio['med_2019'])   

print(f"Base socio-démo prête : {df_socio.shape[0]} communes.")
```

```python
# CHARGEMENT ELECTIONS


# VERIFIER bien CODE NUANCE et pas NUANCE LISTE
df_elec_14 = pd.read_csv(os.path.join(RAW_ELEC_DIR, 'elec_2014_all.csv'), dtype={'Code commune': str})
df_elec_20 = pd.read_csv(os.path.join(RAW_ELEC_DIR, 'elec_2020_all.csv'), dtype={'Code commune': str})


# Renommer si nécessaire (verifier Code Nuance)
df_elec_14 = df_elec_14.rename(columns={'Code commune': 'COM'})
df_elec_20 = df_elec_20.rename(columns={'Code commune': 'COM'})

# DÉFINITION DU BLOC GAUCHE + ÉCOLOS 

BLOC_GAUCHE = [
    'LVECE', 'LVEC', 'LECO',         # Écolos
    'LUG', 'LDVG', 'LSOC', 'LFI',    # Gauche diverse, PS, LFI
    'LGEN', 'LCOM', 'LRDG', 'LGRA', 'LPG' # PCF, Front de gauche, etc.
]

# Idée: mettr à 0 le score si pas de liste de gauche. C'est fait plus bas lors du merge

def calcul_score_bloc(df_elec, nuances_bloc, annee):
    """Calcule la somme des voix du bloc pour chaque commune."""
    # Filtrer les nuances du bloc (gauche dans notre cas)
    mask = df_elec['Code Nuance'].isin(nuances_bloc)
    df_filtered = df_elec[mask].copy()
    
    # Sommer par commune (cas où il y a >=2 listes de gauche)
    scores = df_filtered.groupby('COM')['% Voix/Exp'].sum().reset_index()
    scores.rename(columns={'% Voix/Exp': f'score_gauche_{annee}'}, inplace=True)
    return scores

# appel a la fction, calcul des scores
score_14 = calcul_score_bloc(df_elec_14, BLOC_GAUCHE, '2014')
score_20 = calcul_score_bloc(df_elec_20, BLOC_GAUCHE, '2020')

print("Scores calculés.")
```

```python
# ELECTIONS PRESIDENTIELLES

df_elec_17_pres = pd.read_csv(os.path.join(RAW_ELEC_DIR, 'elec_2017_pres_all.csv'), dtype={'Code commune': str})
df_elec_22_pres = pd.read_csv(os.path.join(RAW_ELEC_DIR, 'elec_2022_pres_all.csv'), dtype={'Code commune': str})

# Définition des blocs 
gauche_2017 = ['MÉLENCHON', 'HAMON', 'POUTOU', 'ARTHAUD']
gauche_2022 = ['MÉLENCHON', 'JADOT', 'HIDALGO', 'ROUSSEL', 'POUTOU', 'ARTHAUD']

# pour 2017 

score_pres_17 = df_elec_17_pres[df_elec_17_pres['Nom'].isin(gauche_2017)] \
                .groupby('Code commune')['% Voix/Exp'].sum() \
                .reset_index(name='score_gauche_pres_2017')

# Pour 2022
score_pres_22 = df_elec_22_pres[df_elec_22_pres['Nom'].isin(gauche_2022)] \
                .groupby('Code commune')['% Voix/Exp'].sum() \
                .reset_index(name='score_gauche_pres_2022')
```

```python
# On part de la base socio-demo (qui contient toutes les communes)
df_final = df_socio.copy()

# On colle les scores 2014
df_final = df_final.merge(score_14, on='COM', how='left')

# On colle les scores 2020
df_final = df_final.merge(score_20, on='COM', how='left')

df_final = df_final.merge(score_pres_17, left_on='COM', right_on='Code commune', how='left')
df_final = df_final.merge(score_pres_22, left_on='COM', right_on='Code commune', how='left')

# CRUCIAL pour la regression: Remplacer les NaN par 0
# Si pas de correspondance, ça veut dire 0 voix pour la gauche
df_final['score_gauche_2014'] = df_final['score_gauche_2014'].fillna(0)
df_final['score_gauche_2020'] = df_final['score_gauche_2020'].fillna(0)

# Calcul du DELTA SCORE (Variable Y)
df_final['delta_score_gauche_municipales'] = df_final['score_gauche_2020'] - df_final['score_gauche_2014']

df_final['delta_score_gauche_pres'] = df_final['score_gauche_pres_2022'] - df_final['score_gauche_pres_2017']

# Log population
# On le calcule ici tant que la population est brute (positive)
df_final['log_pop_2020'] = np.log(df_final['pop_2020'])
df_final['log_pop_2014'] = np.log(df_final['pop_2014'])
df_final['log_pop_2022'] = np.log(df_final['pop_2022'])
df_final['log_pop_2017'] = np.log(df_final['pop_2017'])

# Sélection des colonnes (pour R par exemple, analyse faite en R temporairement)
cols_export = [
    'COM',
    'score_gauche_2014',  
    'score_gauche_2020', 
    'score_gauche_pres_2017',
    'score_gauche_pres_2022',
    'delta_score_gauche_pres',
    'delta_score_gauche_municipales',
    'pop_2014',
    'pop_2017',
    'pop_2020', 
    'pop_2022',
    'log_pop_2014',
    'log_pop_2020',
    'log_pop_2017',
    'log_pop_2022',
    'croissance_pop_14_20',
    'croissance_pop_17_22', 
    'pct_sup_2014',
    'pct_sup_2017',
    'pct_sup_2020',
    'pct_sup_2022',
    'delta_pct_diplome_17_22',
    'delta_pct_diplome_14_20',
     
    'pct_cadres_2014',
    'pct_cadres_2017',
    'pct_cadres_2020',
    'pct_cadres_2022',
    'delta_pct_cadres_17_22',  
    'delta_pct_cadres_14_20', 
    'med_2013',
    'med_2019',  
    'med_2016',
    'log_med_16', 
    'log_med_13',
    'log_med_19' 

]

# On retire les lignes avec des NaN dans les variables explicatives (Socio-demo manquantes)
# faire analyse des données perdues ? 
df_export = df_final[cols_export].dropna()

print(f"Base finale prête : {len(df_export)} communes.")
print(df_export.head())
```

```python
# 1) standardisation des variables pour la regression présidentielles


# Standardisation des variables (X seulement)


df_std_presidentielles = df_export.copy()

# 1. On liste ce qu'on NE VEUT PAS toucher (Variable Cible Y + Identifiants)
# On garde Y naturel pour dire "ça augmente le vote de X points"
cols_excluded = ['COM', 'score_gauche_2014', 'score_gauche_2020', 'delta_score_gauche_pres', 'score_gauche_pres_2017', 'score_gauche_pres_2022', 'delta_score_gauche_municipales']

# 2. On récupère automatiquement toutes les autres variables numériques (les X)
# Cela inclura : croissance_pop, delta_pct_cadres, log_med_13, etc.
x_vars = [c for c in df_std_presidentielles.select_dtypes(include=['float', 'int']).columns if c not in cols_excluded]

print("Variables transformées (X) :", x_vars)
print("Variables conservées naturelles (Y) :", cols_excluded)

# 3. Standardisation (Z-score) des X
for col in x_vars:
    if df_std_presidentielles[col].std() != 0: # Sécurité
        df_std_presidentielles[col] = (df_std_presidentielles[col] - df_std_presidentielles[col].mean()) / df_std_presidentielles[col].std()



# 2) export

output_std_path = os.path.join(MERGED_DATA_DIR, 'data_regression_std_presidentielles.csv')
df_std_presidentielles.to_csv(output_std_path, index=False, sep=',', encoding='utf-8', na_rep='NA')

print(f"Fichier prêt pour l'analyse : {output_std_path}")


```
