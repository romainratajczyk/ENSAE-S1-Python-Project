```python
import pandas as pd 

df_elec_2014 = pd.read_csv('/Users/romain/Desktop/Projets DS/Python-project/analyse/data/elec_2014.csv', dtype={'Code commune':'str'})
df_elec_2020 = pd.read_csv('/Users/romain/Desktop/Projets DS/Python-project/analyse/data/elec_2020.csv', dtype={'Code commune':'str'})

df_elec_2014 = df_elec_2014.rename(columns={'Code commune':'COM','D_elec_2020':'D_elec_2014','Code Nuance':'Code Nuance 2014'})
df_elec_2020 = df_elec_2020.rename(columns={'Code commune':'COM','Nuance Liste':'Code Nuance 2020'})
```

Analyse des données fusionnées (NaN, 0, 1...)

```python
comptage2020 = df_final['D_elec_2020'].value_counts(dropna=False)
print(comptage2020)
comptage2014 = df_final['D_elec_2014'].value_counts(dropna=False)
print(comptage2014)
```

```python
df_clean = df_final.dropna(subset=['D_elec_2014', 'D_elec_2020'])
```

```python
print(df_clean.columns)
```

```python
# creation des variables delta 

# On reconstruit le total "Supérieur" pour 2020 (somme des 3 colonnes)
df_clean['total_diplome_2020'] = (
    df_clean['P20_NSCOL15P_SUP2'] + 
    df_clean['P20_NSCOL15P_SUP34'] + 
    df_clean['P20_NSCOL15P_SUP5']
)



# Delta Vote (Variable Y) : -1 (Perdu), 0 (Stable), +1 (Gagné)
df_clean['delta_elec'] = df_clean['D_elec_2020'] - df_clean['D_elec_2014']

# Delta Population 
df_clean['delta_pop'] = df_clean['P20_POP'] - df_clean['P14_POP']

# Delta Diplômés en %

df_clean['part_diplome_14'] = df_clean['P14_NSCOL15P_SUP'] / df_clean['P14_POP']
df_clean['part_diplome_20'] = df_clean['total_diplome_2020'] / df_clean['P20_POP']

df_clean['delta_pct_diplome'] = df_clean['part_diplome_20'] - df_clean['part_diplome_14']


# Souvent, on prend le stock initial (2013) pour voir si "être riche au début" prédit le changement.

cols_utiles = [
    'COM',           # ID
    'delta_elec',    # Y
    'delta_pop',     # X1
    'delta_pct_diplome', # X2
    'MED13', # X3
              'MED19'        # X4  
]

df_regression = df_clean[cols_utiles].copy()

# Aperçu
print(df_regression.head())
```

```python
# Chemin de sortie
output_path = '/Users/romain/Desktop/Projets DS/Python-project/analyse/data_pour_regression.csv'

# L'exportation propre
df_regression.to_csv(
    output_path, 
    index=False,   # CRUCIAL : Ne pas enregistrer la colonne d'index (0, 1, 2...)
    na_rep='NA',   # CRUCIAL POUR R : Remplacer les vides par "NA"
    sep=',',       # Standard international (virgule)
    encoding='utf-8'
)

print(f"Exporté avec succès vers : {output_path}")
```
